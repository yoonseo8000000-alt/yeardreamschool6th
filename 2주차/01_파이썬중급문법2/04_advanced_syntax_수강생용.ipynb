## 3. sorted – 리스트 정렬

| 방법 | 원본 변경 | 반환값 | 특징 |
|------|----------|--------|------|
| `list.sort()` | 변경 | `None` | 계산 효율 좋음 |
| `sorted(list)` | 유지 | 정렬된 새 리스트 | 원본 보존 가능 |

````python
# ── sort() vs sorted() 비교 ─────────────────────────────────
scores = [85, 92, 78, 96, 73, 88]

# sorted(): 원본 유지, 새 리스트 반환
sorted_scores = sorted(scores)
print('원본    :', scores)          # 변경 없음
print('sorted  :', sorted_scores)   # 정렬된 새 리스트

# sort(): 원본 자체를 변경, None 반환
result = scores.sort()
print('원본(변경후):', scores)      # 원본이 변경됨
print('반환값  :', result)          # None
````

````python
# ── reverse=True: 내림차순 ──────────────────────────────────
scores = [85, 92, 78, 96, 73, 88]

asc  = sorted(scores)               # 오름차순 (기본)
desc = sorted(scores, reverse=True) # 내림차순

print('오름차순:', asc)
print('내림차순:', desc)
````

## 6. `if __name__ == "__main__"`

### 핵심 개념
- Python은 파일을 실행하면 **들여쓰기 없는(최상위) 코드가 모두 실행**됩니다.
- 이 구문을 사용하면 **해당 파일을 직접 실행할 때만** 특정 블록을 실행할 수 있습니다.
- 다른 파일에서 `import`로 불러올 때는 해당 블록이 **실행되지 않습니다**.

```python
# 파일을 직접 실행 시:  __name__ == '__main__'  (True)
# import로 불러올 시:   __name__ == '모듈파일명' (False)
```

````python
# ── __name__ 변수 확인 ───────────────────────────────────────
# Jupyter Notebook에서 실행하면 __name__은 '__main__'
print(f'현재 __name__: {__name__}')

# 항상 True가 됨 (Jupyter는 직접 실행 환경)
if __name__ == '__main__':
    print('이 블록은 직접 실행할 때만 동작합니다.')
````

````python
# ── 실제 사용 예시: 모듈과 메인 코드 분리 ───────────────────

# 함수/클래스 정의 부분 (어디서든 import해서 사용 가능)
def greet(name):
    return f'안녕하세요, {name}님!'

def add(a, b):
    return a + b

# 메인 실행 부분 (직접 실행할 때만 동작)
if __name__ == '__main__':
    # 이 블록은 'python 파일명.py'로 직접 실행할 때만 동작
    # 다른 파일에서 import해도 이 블록은 실행되지 않음
    print(greet('Alice'))
    print(f'3 + 4 = {add(3, 4)}')
    print('메인 프로그램 실행 완료!')
````

````python
# ── sys.exit와 함께 사용하는 패턴 ───────────────────────────
# sys.exit()는 프로그램을 종료할 때 사용하는 함수입니다.
# 종료 코드(exit code)를 반환값으로 전달할 수 있습니다.
#   0      → 정상 종료
#   0 이외  → 오류로 인한 종료
#
# ⚠️  주의: Jupyter Notebook에서 sys.exit()를 직접 호출하면
#          커널(kernel)이 종료되어 작업 내용이 초기화됩니다.
#          .py 파일을 터미널에서 실행할 때만 사용합니다.
#
# 공식 문서: https://docs.python.org/ko/3/library/sys.html#sys.exit

import sys

def compute(a, b):
    """두 수를 더하고 결과를 출력하는 함수"""
    result = a + b
    print(f'  {a} + {b} = {result}')
    return result

def main():
    """
    프로그램의 진입점(entry point) 역할을 하는 함수.
    모든 핵심 로직을 이 함수 안에 작성합니다.
    """
    print('프로그램 시작')
    compute(10, 20)
    compute(100, 200)
    print('프로그램 종료')
    return 0   # 0 = 정상 종료를 sys.exit에 알림


# ── if __name__ == '__main__' 블록 ──────────────────────────
# 이 파일을 직접 실행(python main.py)할 때만 아래 블록이 동작합니다.
# 다른 파일에서 import할 때는 실행되지 않습니다.
#
# [실행 흐름]
#   1. if __name__ == '__main__' 조건 확인
#   2. main() 함수 호출 → 내부 로직 실행
#   3. main()의 반환값(0)을 sys.exit()에 전달
#   4. 운영체제에 종료 코드 0(정상) 반환
#
# sys.exit(main()) 패턴은 아래와 동일합니다:
#   exit_code = main()
#   sys.exit(exit_code)

if __name__ == '__main__':
    sys.exit(main())

# ── Jupyter에서 동작을 확인하고 싶다면 아래처럼 직접 호출 ───
# (sys.exit 없이 main()만 호출)
#
# main()
````

````python
# ── 실행 순서 시각화 ─────────────────────────────────────────
print('=' * 50)
print('실행 순서 확인')
print('=' * 50)

# 1. 함수 정의는 '정의'만 할 뿐, 아직 실행되지 않음
def step_a():
    print('Step A: 함수 A 실행')

def step_b():
    print('Step B: 함수 B 실행')
    step_a()   # A를 호출

# 2. 최상위 코드가 순서대로 실행됨
print('최상위 코드 1')
print('최상위 코드 2')

# 3. __name__ 조건 확인 후 메인 블록 실행
if __name__ == '__main__':
    print('메인 블록 진입')
    step_b()   # B → A 순서로 호출됨
    print('메인 블록 종료')
````

## 7. 종합 실습: 6가지 문법을 모두 활용한 팀 분석

지금까지 배운 f-string, List Comprehension, sorted, enumerate, lambda, `__name__` 을 모두 사용합니다.

````python
# 데이터 준비
employees = [
    {'name': 'Alice',  'team': '개발팀', 'salary': 6500, 'score': 95, 'years': 5},
    {'name': 'Bob',    'team': '기획팀', 'salary': 5200, 'score': 82, 'years': 3},
    {'name': 'Carol',  'team': '영업팀', 'salary': 6100, 'score': 91, 'years': 7},
    {'name': 'Derek',  'team': '개발팀', 'salary': 4800, 'score': 88, 'years': 2},
    {'name': 'Eve',    'team': '분석팀', 'salary': 3600, 'score': 76, 'years': 1},
    {'name': 'Fiona',  'team': '기획팀', 'salary': 4300, 'score': 85, 'years': 4},
    {'name': 'George', 'team': '영업팀', 'salary': 3900, 'score': 79, 'years': 2},
    {'name': 'Helen',  'team': '분석팀', 'salary': 5500, 'score': 93, 'years': 6},
    {'name': 'Ivan',   'team': '개발팀', 'salary': 4200, 'score': 87, 'years': 3},
]

print('=' * 60)
print(' TechCorp Korea - 2024년 상반기 팀원 분석 보고서')
print('=' * 60)

# ── 1. f-string: 요약 통계 출력 ─────────────────────────────
total      = len(employees)
avg_salary = sum(e['salary'] for e in employees) / total
avg_score  = sum(e['score']  for e in employees) / total

print(f'\n  총 직원 수  : {total}명')
print(f'  평균 연봉   : {avg_salary:,.0f}만원')
print(f'  평균 성과점수: {avg_score:.1f}점')

# ── 2. List Comprehension: 조건 필터링 ──────────────────────
high_performers = [e['name'] for e in employees if e['score'] >= 90]
dev_members     = [e['name'] for e in employees if e['team'] == '개발팀']

print(f'\n  성과 우수자(90점+): {", ".join(high_performers)}')
print(f'  개발팀 팀원       : {", ".join(dev_members)}')

# ── 3. sorted + lambda: 성과 점수 순 정렬 ───────────────────
ranked = sorted(employees, key=lambda e: e['score'], reverse=True)

print('\n  성과 순위 (상위 5명):')

# ── 4. enumerate: 순위 번호 부여 ────────────────────────────
medals = {1: '🥇', 2: '🥈', 3: '🥉'}
for rank, emp in enumerate(ranked[:5], start=1):
    medal = medals.get(rank, f'{rank}위')
    # ── f-string: 복합 포맷팅 ──────────────────────────────
    line = (f"    {medal} {emp['name']:<8} [{emp['team']}] "
            f"점수: {emp['score']}점 | 연봉: {emp['salary']:,}만원")
    print(line)

# ── 5. map + lambda: 연봉 인상 계산 ─────────────────────────
# 성과점수 90이상: 5% 인상, 미만: 2% 인상
new_salaries = list(map(
    lambda e: (e['name'], int(e['salary'] * (1.05 if e['score'] >= 90 else 1.02))),
    employees
))

print('\n  연봉 인상 계획:')
for name, new_sal in sorted(new_salaries, key=lambda x: -x[1])[:4]:
    original = next(e['salary'] for e in employees if e['name'] == name)
    diff = new_sal - original
    print(f'    {name:<8}: {original:,}만원 → {new_sal:,}만원 (+{diff}만원)')

# ── 6. __name__ 패턴 ─────────────────────────────────────────
print('\n  [실행 환경]', end=' ')
if __name__ == '__main__':
    print('직접 실행 (Jupyter Notebook)')
else:
    print(f'import 된 모듈 (파일명: {__name__})')

print('\n' + '=' * 60)
print(' 보고서 출력 완료')
print('=' * 60)
````

## 정리: 6가지 심화 문법 한눈에 보기

| 문법 | 형식 | 핵심 포인트 |
|------|------|-------------|
| **f-string** | `f"{변수} {표현식:.2f}"` | Python 3.6+, 간결한 문자열 포맷팅 |
| **List Comprehension** | `[표현식 for x in 시퀀스 if 조건]` | 여러 줄 코드를 한 줄로 |
| **sorted** | `sorted(list, key=lambda, reverse=True)` | 원본 유지, 정렬 기준 자유롭게 |
| **enumerate** | `for i, val in enumerate(list, start=1)` | 인덱스와 값을 동시에 |
| **lambda + map** | `map(lambda x: x*2, list)` | 간단한 변환 함수를 한 줄로 |
| **`__name__`** | `if __name__ == "__main__":` | 직접 실행 vs import 구분 |
